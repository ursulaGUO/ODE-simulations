```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Investigate van der Pol equation given different values of mu

## Preparation

```python
# Define the system of ODE transformed from van der Pol equation
def f(x,t,mu):
    dx = x[1]
    dy = - x[0] + mu * (1 - x[0] * x[0])*x[1]
    return (dx,dy)

# Set mu range and time increment range
mu = np.arange(-10,12,2)
t_range = np.linspace(0, 100, 10000)
```

## Graph van der Pol's phase space given different levels of mu

```python
# Use odeint to solve for van der Pol's solutions, then plot
plt.figure(figsize=(12,8))
for i in mu:
    results = odeint(f, (1,0), t_range, args=(i,))
    x_values = np.array([pair[0] for pair in results])
    x_dot_values = np.array([pair[1] for pair in results])
    plt.plot(x_values, x_dot_values, label=f'mu = {i}')
plt.xlabel('x(t)')
plt.ylabel('x_dot(t)')
plt.legend()
plt.title('van der Pol solutions in phase sapce when x(0)=1, x_dot(0)=0')
plt.show()
```
