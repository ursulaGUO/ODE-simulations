```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

# Investigate Lissajous curves by parametrizing x and y with given a,b,lambda_

## Code functions for plotting

```python
def f(a,b,lambda_,t):
    x = np.cos(a * lambda_ * t)
    y = np.sin(b * lambda_ * t)
    return (x,y)

# Define function to plot Lissajous curve given parameters a, b, and lambda_
def lissajous(a,b,lambda_):
    t_range = np.linspace(0,50,1000)
    results = []
    results2 = [] # results2 doubles the lambda_
    results3 = [] # results3 triples the lambda_

    for i in t_range:
        results.append(f(a,b,lambda_, i))

    
    x_values = np.array([pair[0] for pair in results])
    y_values = np.array([pair[1] for pair in results])

    plt.figure(figsize = (12,8))
    plt.plot(x_values, y_values,label=f'a = {a} ,b = {b},lambda = {lambda_} ')

    plt.xlabel('x(t)')
    plt.ylabel('y(t)')
    plt.legend(loc='upper left')
    plt.title('Lissajous Curve')
    plt.show()
```

## Observe different behavior of the curve given different parameter values

```python
lissajous(1,1,1)
lissajous(1,1,5)
lissajous(1,(1+ np.sqrt(5))/2,1)
lissajous(1,(1+ np.sqrt(5))/2,10)
```
