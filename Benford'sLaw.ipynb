```python
import numpy as np
import matplotlib.pyplot as plt
```

# Investigate Benford's law

## Preparation

```python
# Define function that counts the number of each digit in a number
bins = {int(i): 0 for i in range(1,10)}
def digit_bin(x,bin):
    first = int(str(x)[0])
    bin[first] += 1
    return bin

# Test run for digit_bin
for i in range(0,3): #1,2,4
    num = np.power(2,i)
    bins = digit_bin(num,bins)
print(bins)
```

```python
def plotCountDigit(base,n):
    # Initiate bins to 0
    bins = {int(i): 0 for i in range(1,10)}
    num = 1 
    for i in range(1,100):
        num = base * num
        bins = digit_bin(num,bins)

    # Turn each bin into list
    digits = list(bins.keys())
    counts = list(bins.values())

    # Calculate the average
    average = sum(counts) / len(counts)

    # Create subplot
    fig, ax = plt.subplots()

    # Plot the histogram
    plt.bar(digits, counts, color='orange')
    plt.xlabel('Digit')
    plt.ylabel('Count')
    plt.title(f'Digit Counts Histogram for {base}^n, n in [0,100]')
    ax.axhline(average, color='red', linestyle='--', label='Average')
    plt.legend()
    plt.show()
```

## Counting digits for 2^n, n in [0,100]

```python
plotCountDigit(2,100)
```

## Counting digits for 3^n, n in [0,100]

```python
plotCountDigit(3,100)
```

## Counting digits for 13^n, n in [0,100]

```python
plotCountDigit(13,100)
```
