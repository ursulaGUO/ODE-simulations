```python
import numpy as np
import matplotlib.pyplot as plt
```

# Consider the Lotka Volterra system, plot its phase curves given different x0 and y0

## Preparations

```python
# Define Lotka Volterra system
def f(t,x,y):
    k = 8
    l = 10
    a = 3
    b = 2
    dx = k * x - a * x * y
    dy = - l * y + b * x * y
    return (dx,dy)

# Step
step = 10000
delta = 100 / step
```

## Using Euler's method to test with plotting

```python
# Euler's method
def Euler(x0, y0, t0, delta, step):
    x = x0
    y = y0
    t = t0 
    results = []
    results.append((x,y))
    for i in range(1,step):
        x = x + delta * f(t,x,y)[0]
        y = y + delta * f(t,x,y)[1]
        t = t + delta
        results.append((x,y))
    return results

# test
results1 = Euler(10,10,0,delta,step)
x_values1 = np.array([pair[0] for pair in results1])
y_values1 = np.array([pair[1] for pair in results1])
results2 = Euler(10,15,0,delta,step)
x_values2 = np.array([pair[0] for pair in results2])
y_values2 = np.array([pair[1] for pair in results2])

plt.figure(figsize = (12,8))
plt.plot(x_values1, y_values1, label='x0 = 10, y0 = 10')
plt.plot(x_values2, y_values2, label='x0 = 10, y0 = 15')
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.legend()
plt.show()
```

## Using different step size, plot the Lotka Volterra phase curve given different levels of x0 and y0 with Euler's method

```python
# More detailed plotting with various x0 and y0 pairs
# Step size = 0.1
step = 1000
delta = 100 / step
plt.figure(figsize = (12,8))
for i in range(10,40,10):
    for j in range(10,40,10):
        results = []
        results = Euler(i,j,0,delta,step)
        x_values = np.array([pair[0] for pair in results])
        y_values = np.array([pair[1] for pair in results])
        plt.plot(x_values, y_values,label=f"x0={i}, y0={j}")       
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.title(f"Euler's Method, step size = {delta}")
plt.legend()
plt.show()


# Step size = 0.01
step = 10000
delta = 100 / step
plt.figure(figsize = (12,8))
for i in range(10,40,10):
    for j in range(10,40,10):
        results = []
        results = Euler(i,j,0,delta,step)
        x_values = np.array([pair[0] for pair in results])
        y_values = np.array([pair[1] for pair in results])
        plt.plot(x_values, y_values,label=f"x0={i}, y0={j}")
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.title(f"Euler's Method, step size = {delta}")
plt.legend()
plt.show()
```

## Using Runge Kutta method to test with plotting

```python
# Step size = 0.01
step = 10000
delta = 100 / step

# Runge Kutta Method
def RungeKutta(x0, y0, t0, delta, step):
    x = x0
    y = y0
    t = t0 
    results = []
    results.append((x,y))
    for i in range(1,step):
        x = x + delta * f(t,x,y)[0]
        y = y + delta * f(t,x,y)[1]
        t = t + delta
        k1 = f(t,x,y)
        k2 = f(t + delta / 2, x + delta * k1[0]/2, y + delta * k1[1]/2)
        k3 = f(t + delta / 2, x + delta * k2[0]/2, y + delta * k2[1]/2)
        k4 = f(t + delta / 2, x + delta * k3[0], y + delta * k3[1])
        x = x + delta / 6 * (k1[0] + 2 * k2[0] + 2 * k3[0] + k4[0])
        y = y + delta / 6 * (k1[1] + 2 * k2[1] + 2 * k3[1] + k4[1])
        results.append((x,y))
    return results

# test
results1 = RungeKutta(10,10,0,delta,step)
x_values1 = np.array([pair[0] for pair in results1])
y_values1 = np.array([pair[1] for pair in results1])
results2 = RungeKutta(10,15,0,delta,step)
x_values2 = np.array([pair[0] for pair in results2])
y_values2 = np.array([pair[1] for pair in results2])

plt.figure(figsize = (12,8))
plt.plot(x_values1, y_values1, label='x0 = 10, y0 = 10')
plt.plot(x_values2, y_values2, label='x0 = 10, y0 = 15')
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.legend()
plt.show()
```

## Using different step size, plot the Lotka Volterra phase curve given different levels of x0 and y0 with Runge Kutta method

```python
# Step size = 0.01
step = 10000
delta = 100 / step
plt.figure(figsize = (12,8))
for i in range(10,40,10):
    for j in range(10,40,10):
        results = []
        results = RungeKutta(i,j,0,delta,step)
        x_values = np.array([pair[0] for pair in results])
        y_values = np.array([pair[1] for pair in results])
        plt.plot(x_values, y_values,label=f"x0={i}, y0={j}")
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.title(f"Runge Kutta Method, step size = {delta}")
plt.legend()
plt.show()

# Step size = 0.001
step = 100000
delta = 100 / step
plt.figure(figsize = (12,8))
for i in range(10,40,10):
    for j in range(10,40,10):
        results = []
        results = RungeKutta(i,j,0,delta,step)
        x_values = np.array([pair[0] for pair in results])
        y_values = np.array([pair[1] for pair in results])
        plt.plot(x_values, y_values,label=f"x0={i}, y0={j}")
plt.xlabel('X Population')
plt.ylabel('Y Population')
plt.title(f"Runge Kutta Method, step size = {delta}")
plt.legend()
plt.show()
```
