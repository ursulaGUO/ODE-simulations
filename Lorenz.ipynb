```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

# Investigate Lorenz system through graphing in 3d space

## Preparationo and test run

```python
def f(state, t, sigma, ro, beta):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * (ro - z) - y
    dz = x * y - beta * z
    return [dx, dy, dz]

t_range = np.linspace(0, 50, 1000)

# Parameters
sigma = 10
beta = 8/3
ro = 28

# Test run start state
start = (1, 1, 1)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

results = odeint(f, start, t_range, args=(sigma, ro, beta))

x_values = np.array([pair[0] for pair in results])
y_values = np.array([pair[1] for pair in results])
z_values = np.array([pair[2] for pair in results])

ax.plot(x_values, y_values, z_values, label=f'x0, y0, z0 = ({start[0]}, {start[1]}, {start[2]})')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Lorenz system phase curves')
plt.legend()
plt.show()
```

## lorenz function to generate phase curve plot in 3d

```python
# Lorenz function that inputs the parameters, and generate 5 different graphs given 5 different inital state
def lorenz(sigma, ro, beta, start):
    t_range = np.linspace(0, 50, 10000)

    fig = plt.figure(figsize=(12,8))
    ax = fig.add_subplot(111, projection='3d')  

    for i in start: 
        results = odeint(f, i, t_range, args=(sigma, ro, beta))

        x_values = np.array([pair[0] for pair in results])
        y_values = np.array([pair[1] for pair in results])
        z_values = np.array([pair[2] for pair in results])

        ax.plot(x_values, y_values, z_values, label=f'x0, y0, z0 = ({i[0]}, {i[1]}, {i[2]})')

    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title('Lorenz System Phase Curves')
    plt.legend()
    plt.show()

# First pair of starting point (1,1,1) and (-1,-1,-1)
start = [[1, 0, 0],[0.5,0,0], [-1,0,0]]
lorenz(10,28,8/3, start)

# Second pair of starting point (1,1,1) and (-1,-1,-1)
start = [[100,100,100],[40,40,40],[10,10,10],[5,5,5]]
lorenz(10,28,8/3, start)
```

## lorenzXIntCurve to generate the intergral curve of x variable

```python
# Plotting x-coordinate of trajectory over time
def lorenzXIntCurve(sigma, ro, beta, start):
    t_range = np.linspace(0, 50, 10000)

    plt.figure(figsize=(12,8))
    
    for i in start: 
        results = odeint(f, i, t_range, args=(sigma, ro, beta))

        x_values = np.array([pair[0] for pair in results])

        plt.plot(t_range, x_values, label=f'x0, y0, z0 = ({i[0]}, {i[1]}, {i[2]})')

    plt.xlabel('t')
    plt.ylabel('x')
    plt.title('Integral curve in the x-coordinate')
    plt.legend(loc='upper right')
    plt.show()

# First pair of starting point (1,1,1) and (-1,-1,-1)
start = [[1, 0, 0],[0.5,0,0], [-1,0,0]]
lorenzXIntCurve(10,28,8/3, start)

start = [[100,100,100],[40,40,40],[10,10,10],[5,5,5]]
lorenzXIntCurve(10,28,8/3, start)
```
